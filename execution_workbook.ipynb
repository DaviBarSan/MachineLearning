```python
# Generate a bidimensional dataset with class control
X, y = make_classification(
    n_samples=100,              # Number of instances (rows)
    n_classes=4,                # Number of classes
    n_informative=3,            # Number of informative features
    n_clusters_per_class=2,     # Number of clusters per class
    weights=[0.3, 0.3, 0.4],    # Class proportions (imbalanced dataset)
    class_sep=10.5,              # Separation between classes (higher = more separable)
    flip_y=0.001,                 # Level of noise (mislabeling fraction)
    random_state=42              # Reproducibility
)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Synthetic Classification Dataset")
plt.show()
```
